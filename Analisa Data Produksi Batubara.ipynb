# 1. Deskripsi Data

Dataset ini berisi mengenai produksi batu bara di Indonesia dari tahun 2003 - 2010

Penjelasan mengenai Variabel pada Dataset ini:

    tipe_entitas: Tipe Entitas
    nama_entitas: Nama Entitas
    tahun: Tahun
    nilai: Nilai (ton)

Sumber data :
https://data.go.id/dataset/produksi-batu-bara

```python
import pandas as pd
```

```python
data_produksi = pd.read_csv("processed-produksi-batubara-2003-2010.csv", delimiter = ",", encoding = "utf-8")
```

# 2. Statistik Data

```python
jumlah = data_produksi.nilai.sum()
rata = data_produksi.nilai.mean()
mi = data_produksi.nilai.min()
ma = data_produksi.nilai.max()

print("Jumlah Produksi =", jumlah, "ton")
print("Rata-rata Produksi =", rata, "ton")
print("Minimum Produksi =", mi, "ton")
print("Maksimum Produksi =", ma, "ton")
```

```python
data_produksi.nilai.describe()
```

```python
data_produksi.groupby(["tipe_entitas",]).sum()["nilai"]
```

```python
data_produksi.groupby(["nama_entitas", "tahun"]).sum()["nilai"]
```

# 3. Visualisasi Data

```python
ax = data_produksi.groupby('nama_entitas').sum()['nilai']
ax = ax.plot(kind='bar', title ="Jumlah Produksi Batubara dari Tahun 2003 - 2010", figsize=(17,5), legend=True, fontsize=12)
ax.set_xlabel("Nama Perusahaan", fontsize=12)
ax.set_ylabel("Nilai/Jumlah", fontsize=12)
```

```python
ax = data_produksi.groupby('tipe_entitas').sum()['nilai']
ax = ax.plot(kind='bar', title ="Jumlah Produksi Batubara dari Tahun 2003 - 2010", figsize=(17,5), legend=True, fontsize=12)
ax.set_xlabel("Tipe Entitas", fontsize=12)
ax.set_ylabel("Nilai/Jumlah", fontsize=12)
```

# 4. Narasi Analisis

Berdasarkan visualisasi data di atas, disimpulkan bahwa tipe entitas Kontraktor memproduksi batubara paling tinggi dari tahun 2003 - 2010, yaitu sebesar 1.100.134.000 ton, diikuti BUMN sebesar 67.980.120 ton dan Kuasa Pertambangan Swasta sebesar 23.490.350 ton. PT. Adaro Indonesia menyumbang produksi tertinggi pada sisi Kontraktor sebanyak 234.382.800 ton, dimana produksi tertinggi terjadi pada tahun 2009 sebesar 40.590.189 ton.
