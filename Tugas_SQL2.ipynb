```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

# One to many relationship

1. Report the account representative for each customer

```python
run("""
SELECT
    employeeNumber,
    CONCAT(T1.lastName, ", ", T1.firstName) AS EmployeeName,
    customerName
FROM
    employees T1
        INNER JOIN
    customers T2 ON T1.employeeNumber = T2.salesRepEmployeeNumber
ORDER BY
    employeeName
""")
```

2. Reports total payments for Atelier graphique

```python
run("""
SELECT
    customerName,
    SUM(amount)
FROM
    customers
        INNER JOIN
    payments USING (customerNumber)
WHERE
    customerName = "Atelier graphique"
""")
```

3. Report the total payments by date

```python
run("""
SELECT
    paymentDate,
    SUM(amount) AS TotalPayments
FROM
    payments
GROUP BY
    paymentDate
""")
```

4. Report the Products that have not been sold

```python
run("""
SELECT
    productName,
    orderNumber
FROM
    products a
        LEFT JOIN
    orderdetails b ON a.productCode = b.productCode
WHERE
    orderNumber is NULL
""")
```

5. List the amount paid by each customer

```python
run("""
SELECT
   customerName,
   sum(amount)
FROM
    customers
        INNER JOIN
    payments USING (customerNumber)
GROUP BY
    customerName
""")
```

6. How many orders have been placed by Herkku Gifts?

```python
run("""
SELECT
    customerName,
    COUNT(orderNumber) AS TotalOrder
FROM
    customers
        INNER JOIN
    orders USING (customerNumber)
WHERE
    customerName = "Herkku Gifts"
""")
```

7. Who are the employees in Boston?

```python
run("""
SELECT
    city,
    CONCAT(firstName, ", ", lastName) AS EmployeeName
FROM
    offices
        INNER JOIN
    employees USING (officeCode)
WHERE
    city = "Boston"    
""")
```

8. Report those payments greater than $100.000. Sort the report so the customer who made the highest payment appears first.

```python
run("""
SELECT
   customerName,
   SUM(amount) AS Total
FROM
    customers
        INNER JOIN
    payments USING (customerNumber)
WHERE
    amount > 100000
GROUP BY
    customerName
ORDER BY
    Total DESC
""")
```

9. List the value of 'On Hold' orders.

```python
run("""
SELECT
    *
FROM
    orders
WHERE
    status ="On Hold"
""")
```

10. Report the number of orders 'On Hold' for each customer.

```python
run("""
SELECT
    customerName,
    status,
    COUNT(b.customerNumber) AS TotalOrders
FROM
    customers a
        INNER JOIN
    orders b ON a.customerNumber = b.customerNumber
WHERE
    status = "On Hold"
GROUP BY
    customerName
""")
```

# Many to many relationship

1. List products sold by order date.

```python
run("""
SELECT 
    orderDate,
    productCode,
    productName
FROM
    products
        INNER JOIN
    orderdetails USING (productCode)
        INNER JOIN
    orders USING (orderNumber)
ORDER BY
    orderDate, productCode
""")
```

2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
run ("""
SELECT 
    orderDate,
    orderNumber,
    productName
FROM
    orderdetails
        INNER JOIN
    orders USING (orderNumber)
        INNER JOIN
    products USING (productCode)
WHERE
    productName LIKE "%1940 Ford Pickup Truck%"
ORDER BY
    orderDate DESC
""")
```

3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
run ("""
SELECT 
    customerName, 
    orderNumber,
    SUM(quantityOrdered*priceEach) as Payment
FROM
    customers
        INNER JOIN
    orders USING (customerNumber)
        INNER JOIN
    orderdetails USING (orderNumber)
GROUP BY 
    orderNumber    
HAVING SUM(quantityOrdered*priceEach) > 25000
ORDER BY
    Payment
""")
```

4. Are there any products that appear on all orders?

```python
run ("""
SELECT 
    a.productName, 
    COUNT(b.productCode) as TotalOrder
FROM
    orderdetails b 
        JOIN 
    products a USING (productcode) 
        JOIN 
    orders c USING (orderNumber)
GROUP BY
    productCode 
HAVING TotalOrder = (SELECT COUNT(*) from orders)
ORDER BY
    TotalOrder DESC
""")
```

5. List the names of products sold at less than 80% of the MSRP.

```python
run ("""
SELECT
    productName,
    priceEach,
    MSRP
FROM
    products
        INNER JOIN 
    orderdetails USING (productCode)
        INNER JOIN
    orders USING (orderNumber)
WHERE
    priceEach < MSRP*0.8
""")
```

6. Reports those products that have been sold with a markup of 100% or more (i.e.,  the priceEach is at least twice the buyPrice)

```python
run ("""
SELECT
    productName,
    buyPrice,
    priceEach
FROM
    products
        INNER JOIN 
    orderdetails USING (productCode)
WHERE
    priceEach-buyPrice > buyprice
""")
```

7. List the products ordered on a Monday.

```python
run ("""
SELECT
    orderDate,
    DAYNAME(orderDate) AS Day,
    productName
FROM
    products
        INNER JOIN 
    orderdetails USING (productCode)
        INNER JOIN
    orders USING (orderNumber)
WHERE
    DAYNAME(orderDate) = "Monday"
""")
```

8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
run("""
SELECT 
    productName,
    SUM(quantityOrdered) AS Quantities,
    status
FROM 
    orders
        INNER JOIN
    orderdetails USING (orderNumber)
        INNER JOIN
    products USING (productCode)
WHERE
    status = "On Hold"
GROUP BY
    productName
""")
```
